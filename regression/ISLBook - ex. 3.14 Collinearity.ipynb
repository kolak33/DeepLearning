```python
%config IPCompleter.greedy=True
import numpy as np
```

```python
np.random.seed(1)
x1 = np.random.uniform(size = 100)
x2 = 0.5 * x1 + np.random.normal(loc = 0, scale = 1, size = 100) / 10
y = -2 + 2 * x1 + 0.3 * x2 + np.random.normal(loc = 0, scale = 1, size = 100)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, x2)
plt.xlabel('x1', fontsize=18)
plt.ylabel('x2', fontsize=16)
plt.legend()

np.corrcoef(x1, x2)
```

```python
from patsy import dmatrices, dmatrix
import statsmodels.api as sm
import pandas as pd

Xdmatrix = dmatrix("x1 + x2 + 1", {"x1" : x1, "x2" : x2})
res = sm.OLS(y, Xdmatrix).fit()
res.summary()
```


# y ~ x1 + 1

```python
Xdmatrix = dmatrix("x1 + 1", {"x1" : x1, "x2" : x2})
res = sm.OLS(y, Xdmatrix).fit()
res.summary()
```


# y ~ x2 + 1

```python
Xdmatrix = dmatrix("x2 + 1", {"x1" : x1, "x2" : x2})
res = sm.OLS(y, Xdmatrix).fit()
res.summary()
```

```python
x1 = np.append(x1, [0.1])
x2 = np.append(x2, [0.8])
y = np.append(y, [6])
```

```python
from patsy import dmatrices, dmatrix
import statsmodels.api as sm
import pandas as pd

Xdmatrix = dmatrix("x1 + x2 + 1", {"x1" : x1, "x2" : x2})
res1 = sm.OLS(y, Xdmatrix).fit()
print(res1.summary())
```

```python
Xdmatrix = dmatrix("x1 + 1", {"x1" : x1, "x2" : x2})
res2 = sm.OLS(y, Xdmatrix).fit()
res2.summary()
```

```python
Xdmatrix = dmatrix("x2 + 1", {"x1" : x1, "x2" : x2})
res3 = sm.OLS(y, Xdmatrix).fit()
res3.summary()
```

```python
print('pred 1: ', res1.params[0] + res1.params[1] * x1[-1] * res1.params[2] * x2[-1])
print('pred 2: ', res2.params[0] + res2.params[1] * x1[-1])
print('pred 3: ', res3.params[0] + res3.params[1] * x2[-1])
```
